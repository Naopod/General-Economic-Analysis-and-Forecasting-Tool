```python
import wbdata
import pandas as pd
import datetime

def fetch_data(indicators, date_range):
    return wbdata.get_dataframe(
        indicators,
        country=['USA'],
        date=date_range,
    )

indicators = {
        "NY.GDP.MKTP.CD": "GDP (current US$)",
        "NY.GDP.PCAP.CD": "GDP per capita (current US$)",
        "NY.GDP.DEFL.KD.ZG": "Inflation, GDP deflator (annual %)",
        "NY.GNS.ICTR.CD": "Gross savings (current US$)",
        "BM.GSR.GNFS.CD": "Imports of goods and services (BoP, current US$)",
        "BN.KLT.DINV.CD": "Foreign direct investment, net (BoP, current US$)",
        "BX.GSR.GNFS.CD": "Exports of goods and services (BoP, current US$)",
        "FP.CPI.TOTL.ZG": "Inflation, consumer prices (annual %)"
}

start_date = datetime.datetime(1960, 1, 1) 
end_date = datetime.datetime.now()  

df = fetch_data(
    indicators=indicators,
    date_range=(str(start_date), str(end_date))
)

df.dropna(axis=0, inplace=True)
```

```python
def plot_gdp_acf_pacf(data, title_suffix):
    """Plot ACF and PACF for given GDP data."""
    fig, (ax, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(10, 2))
    plot_acf(data['gdp_growth'].dropna(), lags=20, title=f"GDP Growth ACF ({title_suffix})", ax=ax)
    ax.grid(True)
    plot_pacf(data['gdp_growth'].dropna(), lags=20, title=f"GDP Growth PACF ({title_suffix})", ax=ax1)
    ax1.grid(True)
    plt.tight_layout()
    plt.show()

plot_gdp_acf_pacf(gdp, "including COVID")
plot_gdp_acf_pacf(gdp[gdp['year'] < 2020], "excluding COVID")
```

```python
orders = [(1,0,0), (2,0,0), (3,0,0), (1,0,1), (1,0,2), (1,0,3), (2,0,1), (2,0,2), (2,0,3)]
models, aic_values, bic_values = [], [], []

for order in orders:
    model = ARIMA(gdp['gdp_growth'].dropna(), order=order).fit()
    models.append(f'ARMA({order[0]},{order[2]})')
    aic_values.append(model.aic)
    bic_values.append(model.bic)

results = pd.DataFrame({'Models': models, 'AIC': aic_values, 'BIC': bic_values})
print(tabulate(results, headers='keys', tablefmt='grid'))
```

```python
def create_sequences(input_data, tw):
    """
    Transform a time series into a prediction dataset
    X : our feature
    y : prediction
    tw : training window
    """
    X, y = [], []
    L = len(input_data)
    for i in range(L - tw):
        feature = input_data[i:i+tw]
        target = input_data[i+tw]
        X.append(feature)
        y.append(target)
    return torch.tensor(X, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)
```

```python
gdp_raw = gdp[['Gross domestic product']].values.astype('float32')
scaler = MinMaxScaler()
gdp_normalized = scaler.fit_transform(gdp_raw)
```

```python
plt.figure(figsize=(15, 3))
plt.plot(gdp_normalized)   
plt.title("GDP Over Time")
plt.xlabel("Quarters")
plt.ylabel("GDP")
plt.grid(True)
plt.show()
```

```python
train_size = int(len(gdp_normalized)*0.70)

train, test = gdp_normalized[:train_size], gdp_normalized[train_size:]

lookback = 3
X_train, y_train = create_sequences(train, lookback)
X_test, y_test = create_sequences(test, lookback)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
class LSTM(torch.nn.Module):
    def __init__(self, input_dim=1, hidden_dim=15, output_dim=1, layer_num=1):
        super().__init__()
        self.lstm = torch.nn.LSTM(input_dim, hidden_dim, layer_num, batch_first=True)
        self.linear = torch.nn.Linear(hidden_dim, output_dim)
    def forward(self, input_seq):
        lstm_out, _ = self.lstm(input_seq)
        
        last_out = lstm_out[:, -1, :]  
        
        output = self.linear(last_out)  
        
        return output
```

```python
tw = 12

model = LSTM()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_function = torch.nn.MSELoss()
loader = torch.utils.data.DataLoader(torch.utils.data.TensorDataset(X_train, y_train), shuffle=True, batch_size=8)
```

```python
epochs_n = 2000
train_rmse_list = []
test_rmse_list = []
increasing_count = 0
prev_test_loss = np.inf


for epoch in range(epochs_n):
    model.train()
    for x_batch, y_batch in loader:
        model.zero_grad()
        pred_y = model(x_batch)
        loss = loss_function(pred_y, y_batch)
        loss.backward()
        optimizer.step()

    if epoch % 100 == 0:
        with torch.no_grad():
            model.eval()
            pred_y = model(X_train)
            train_loss = np.sqrt(loss_function(pred_y, y_train))
            train_rmse_list.append(train_loss.item())
            
            pred_test_y = model(X_test)
            test_loss = np.sqrt(loss_function(pred_test_y, y_test))
            test_rmse_list.append(test_loss.item())
                    
            print(f"Epoch {epoch}: Training RMSE: {train_loss:.4f}, Testing RMSE: {test_loss:.4f}")

            if test_loss > prev_test_loss:
                increasing_count += 1
            else:
                increasing_count = 0

            prev_test_loss = test_loss

            if increasing_count == 2:
                print("Stopping early due to increasing test RMSE.")
                break
```

```python
epochs_to_plot = range(0, len(train_rmse_list) * 100, 100) 

plt.figure(figsize=(10, 6))
plt.plot(epochs_to_plot, train_rmse_list, label='Training RMSE', color='blue')
plt.plot(epochs_to_plot, test_rmse_list, label='Testing RMSE', color='orange')
plt.xlabel('Epochs')
plt.ylabel('RMSE')
plt.title('Training and Testing RMSE over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
with torch.no_grad():
    train_plot = np.ones_like(gdp_normalized) * np.nan
    test_plot = np.ones_like(gdp_normalized) * np.nan

    y_pred_train = model(X_train).view(-1).numpy()  
    train_plot[lookback:train_size] = y_pred_train.reshape(train_plot[lookback:train_size].shape)

    y_pred_test = model(X_test).view(-1).numpy()  
    test_plot[train_size + lookback:] = y_pred_test.reshape(test_plot[train_size + lookback:].shape)

plt.figure(figsize=(10, 6))
plt.plot(gdp_normalized, label='Actual Data', color='blue')
plt.plot(train_plot, label='Train Predictions', color='red')
plt.plot(test_plot, label='Test Predictions', color='green')
plt.legend()
plt.title("LSTM Predictions vs Actual Data")
plt.xlabel("Time")
plt.ylabel("Normalized GDP")
plt.show()
```

```python
df = pd.read_excel('../data/National Income and Product Accounts.xlsx', skiprows=5)
df = df.iloc[:,1:]
df.index = df.iloc[:,0]

data_1 = df.copy()
cols = list(data_1.columns)
for i in range(1, len(cols)):
    if 'Unnamed' in cols[i]:
        cols[i] = cols[i-1]
data_1.columns = cols

data_1.columns = [str(col) + str(sub_col) if pd.notna(sub_col) else str(col) for col, sub_col in zip(data_1.columns, data_1.iloc[0,:])]
data_1 = data_1.iloc[1:11,1:]

data_1.index = data_1.index.str.strip()
data_1 = data_1.transpose()
data_1["year"] = pd.to_numeric(data_1.index.str[:4])
data_1["yr_q"] = data_1.index
data_1.index = pd.to_datetime(data_1.index)

gdp = data_1[["Gross domestic product", "year"]].dropna()
gdp["gdp_growth"] = gdp["Gross domestic product"].pct_change()
gdp_growth_log = [np.log(gdp["Gross domestic product"].values[i+1] / gdp["Gross domestic product"].values[i]) for i in range(len(gdp)-1)]
gdp_growth_log = [None] + gdp_growth_log

gdp["gdp_growth_log"] = gdp_growth_log

gdp_raw = gdp[['Gross domestic product']].values.astype('float32')
scaler = MinMaxScaler()
gdp_normalized = scaler.fit_transform(gdp_raw)

def split_sequence(sequence, n_steps_in, n_steps_out):
	X, y = list(), list()
	for i in range(len(sequence)):
		end_ix = i + n_steps_in
		out_end_ix = end_ix + n_steps_out
		if out_end_ix > len(sequence):
			break
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix:out_end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return torch.tensor(X, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)

train_size = int(len(gdp_normalized)*0.70)

train, test = gdp_normalized[:train_size], gdp_normalized[train_size:]

lookback = 3
X_train, y_train = split_sequence(train, lookback, n_steps_out=2)
X_test, y_test = split_sequence(test, lookback, n_steps_out=2)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
class EncoderDecoderLSTM(nn.Module):
    def __init__(self, input_dim=1, hidden_dim=100, n_steps_out=2):
        super(EncoderDecoderLSTM, self).__init__()
        self.encoder_lstm = nn.LSTM(input_dim, hidden_dim, batch_first=True)
        self.decoder_lstm = nn.LSTM(hidden_dim, hidden_dim, batch_first=True)
        self.repeat_vector = n_steps_out
        self.output_layer = nn.Linear(hidden_dim, 1)

    def forward(self, x):
        _, (hidden, cell) = self.encoder_lstm(x)
        
        repeated_vector = hidden.permute(1, 0, 2).repeat(1, self.repeat_vector, 1)
        
        decoder_output, _ = self.decoder_lstm(repeated_vector, (hidden, cell))
        
        output = self.output_layer(decoder_output)
        return output

model = EncoderDecoderLSTM()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_function = torch.nn.MSELoss()
loader = torch.utils.data.DataLoader(torch.utils.data.TensorDataset(X_train, y_train), shuffle=True, batch_size=8)

print(model)
```

```python
epochs_n = 2000
train_rmse_list = []
test_rmse_list = []
increasing_count = 0
prev_test_loss = np.inf

for epoch in range(epochs_n):
    model.train()
    for x_batch, y_batch in loader:
        model.zero_grad()
        pred_y = model(x_batch)
        loss = loss_function(pred_y, y_batch)
        loss.backward()
        optimizer.step()

    if epoch % 100 == 0:
        with torch.no_grad():
            model.eval()
            pred_y = model(X_train)
            train_loss = np.sqrt(loss_function(pred_y, y_train))
            train_rmse_list.append(train_loss.item())
            
            pred_test_y = model(X_test)
            test_loss = np.sqrt(loss_function(pred_test_y, y_test))
            test_rmse_list.append(test_loss.item())
                    
            print(f"Epoch {epoch}: Training RMSE: {train_loss:.4f}, Testing RMSE: {test_loss:.4f}")

            if test_loss > prev_test_loss:
                increasing_count += 1
            else:
                increasing_count = 0

            prev_test_loss = test_loss

            if increasing_count == 2:
                print("Stopping early due to increasing test RMSE.")
                break
```

```python
def split_sequences(sequences, n_steps_in, n_steps_out):
	X, y = list(), list()
	for i in range(len(sequences)):
		end_ix = i + n_steps_in
		out_end_ix = end_ix + n_steps_out
		if out_end_ix > len(sequences):
			break
		seq_x, seq_y = sequences[i:end_ix, :], sequences[end_ix:out_end_ix, :]
		X.append(seq_x)
		y.append(seq_y)
	return torch.tensor(X, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)
```

```python
in_gdp = in_gdp.reshape((len(in_gdp), 1))
in_pcec = in_pcec.reshape((len(in_pcec), 1))

scaler_gdp = MinMaxScaler()
scaler_pcec = MinMaxScaler()

in_gdp_normalized = scaler_gdp.fit_transform(in_gdp)
in_pcec_normalized = scaler_pcec.fit_transform(in_pcec)

dataset = np.hstack((in_gdp_normalized, in_pcec_normalized))
```

```python
class EncoderDecoderLSTM(nn.Module):
    def __init__(self, n_steps_out, n_features, hidden_size=200):
        super(EncoderDecoderLSTM, self).__init__()
        self.hidden_size = hidden_size
        self.n_steps_out = n_steps_out
        self.n_features = n_features

        self.lstm1 = nn.LSTM(input_size=n_features, hidden_size=hidden_size, batch_first=True)
        self.relu = nn.ReLU()

        self.repeat_vector = nn.Linear(hidden_size, n_steps_out * hidden_size)

        self.lstm2 = nn.LSTM(input_size=hidden_size, hidden_size=hidden_size, batch_first=True)

        self.time_distributed = nn.Linear(hidden_size, n_features)

    def forward(self, x):
        batch_size = x.size(0)

        out, (hidden, _) = self.lstm1(x)
        out = self.relu(out[:, -1, :]) 

        out = self.repeat_vector(out)
        out = out.view(batch_size, self.n_steps_out, self.hidden_size)  

        out, _ = self.lstm2(out)

        out = self.time_distributed(out)

        return out
    
model = EncoderDecoderLSTM(n_steps_out=n_steps_out, n_features=n_features)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_function = torch.nn.MSELoss()
loader = torch.utils.data.DataLoader(torch.utils.data.TensorDataset(X_train, y_train), shuffle=True, batch_size=8)

print(model)
```

```python
class LSTM(nn.Module):
    def __init__(self, num_classes, input_size, hidden_size, num_layers):
        super(LSTM, self).__init__()
        self.num_classes = num_classes 
        self.num_layers = num_layers  
        self.input_size = input_size  
        self.hidden_size = hidden_size  
        self.lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size,
                            num_layers=num_layers, batch_first=True, dropout=0.2)  #
        self.fc_1 = nn.Linear(hidden_size, 128)  
        self.fc_2 = nn.Linear(128, num_classes)  
        self.relu = nn.ReLU()

    def forward(self, x):
        h_0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c_0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        output, (hn, cn) = self.lstm(x, (h_0, c_0))  
        
        hn = hn[-1]  
        out = self.relu(hn)
        out = self.fc_1(out)  
        out = self.relu(out)  
        out = self.fc_2(out)  
        return out
```

```python
def training_loop(n_epochs, lstm, optimiser, loss_fn, X_train, y_train, X_test, y_test):
    for epoch in range(n_epochs):
        lstm.train()
        outputs = lstm(X_train)
        optimiser.zero_grad()
        loss = loss_fn(outputs, y_train)
        loss.backward()
        optimiser.step()
        
        if epoch % 100 == 0:
            lstm.eval()
            with torch.no_grad():
                test_preds = lstm(X_test)
                test_loss = loss_fn(test_preds, y_test)
            print(f"Epoch: {epoch}, Train Loss: {loss.item():.5f}, Test Loss: {test_loss.item():.5f}")
```
